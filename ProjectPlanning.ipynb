**3. Exploratory Data Analysis and Visualization**

```r
# Demonstrate the dataset can be loaded
sessions
players
```

```r
## Wrangling

# players is tidy data
# sessions is not
# start_time/end_time and original_start_time/original_end_time are
# two columns representing the same thing in two different formats

# Convert to datetime as this is the most versitile form of the data
# then remove all of the duplicate column representations
sessions_tidy <- sessions |> mutate(start_datetime = dmy_hm(start_time),
                            end_datetime = dmy_hm(end_time)) |> 
                            select(hashedEmail, start_datetime, end_datetime)
# show that this can be loaded
sessions_tidy
```

```r
## Compute Mean Values of players.csv

players_mean_values <- players |> select(played_hours, Age) |>
                        map_dfr(mean, na.rm = TRUE)
players_mean_values
```

```r
## Exploratory Visualizations
options(repr.plot.width = 10, repr.plot.height = 6) 

players_scatter <- players |> 
        ggplot(aes(x = Age, y = played_hours, color = gender)) +
        geom_point(aes(color = gender)) + 
        labs(title = "Hours of Minecraft Played vs Age of Player",
            x = "Age [Years]", 
            y = "Hours Played [Hours]", 
            color = "Gender") +
        theme(text=element_text(size = 20))
        
players_scatter

players_box <- players |> ggplot(aes(subscribe, played_hours, fill = subscribe)) +
  geom_boxplot() +
  labs(
    title = "Hours of Minecraft Played Based On Subscription Status",
    x = "Subscribed?",
    y = "Hours Played [Hours]",
    fill = "Subscribed") +
  theme(text = element_text(size = 20))

players_box
```

From the scatterplot, we can see that the average play time of the players tends to decrease once they are over the age of 30. This suggests that younger people are more likely to play for longer hours. In addition, it appears that the players with the highest played hours are male and female. However, given that there are many points at or around 0, there are no definitive conclusions that can be drawn at this moment. With the boxplot, we can see that the played hours increases by several fold if the player is subscribed. 

**Methods and Plan**
Linear regression is an appropriate method for predicting hours_played because the target variable is continuous, and with only three predictors Age, gender, and subcription, the model is simple, interpretable, and provides a useful baseline. To apply linear regression, several assumptions should be reasonably satisfied. the relationship between predictors and the target should be linear, observations must be independent, the residuals should have constant variance and roughly follow a normal distribution, and the predictors should not be highly correlated.
With regards to the limitations, one must consider the data that is being provided. The variance in the data is quite high and seems to contain quite a few outliers. In addition, there are many categories in the gender variable that could impact the meaningfulness of the data. Specifically "Prefer Not To Say" as a gender is not highly useful as the player could be any of the given genders. In terms of data processing, the categorical variables should be encoded, and the dataset should be split into training and testing sets, likely with a 75/25 split. K-fold cross-validation will also be applied on the training set to improve generalization and detect overfitting.
